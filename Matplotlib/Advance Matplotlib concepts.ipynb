```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

### Colored Scatterplots. 

```python
iris_df = sns.load_dataset('iris')
iris_df.sample(5)
```

```python
iris_df['species'] = iris_df['species'].replace({'setosa' : 0, 'versicolor' : 1, 'virginica' : 2})
```

```python
# cmap and alpha
plt.scatter(iris_df['sepal_length'], iris_df['petal_length'], c = iris_df['species'], cmap='jet', alpha=0.6)

plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()

plt.show()
```

### Plot size

```python
plt.figure(figsize=(12, 6))

# cmap and alpha
plt.scatter(iris_df['sepal_length'], iris_df['petal_length'], c = iris_df['species'], cmap='jet', alpha=0.6)

plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()

plt.show()
```

### Annotations

```python
batsman = pd.read_csv('../Datasets/batter.csv')
```

```python
sample_df = batsman.head(100).sample(25, random_state = 5)
```

```python
sample_df
```

```python
plt.figure(figsize = (15, 8))

plt.scatter(sample_df['avg'], sample_df['strike_rate'], s = sample_df['runs'], marker='o')

for i in range(0, sample_df.shape[0]): 
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

### Horizontal and Vertical lines.

```python
plt.figure(figsize = (15, 8))

plt.scatter(sample_df['avg'], sample_df['strike_rate'], s = sample_df['runs'], marker='o')

plt.axhline(130, color = 'red')
plt.axhline(140, color = 'green')
plt.axvline(30, color = 'red')
plt.axvline(35, color = 'green')

for i in range(0, sample_df.shape[0]): 
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

### Subplots

```python
# how to make a subplots. 
fig, ax = plt.subplots(figsize = (10, 4))

ax.scatter(batsman['avg'], batsman['strike_rate'], color = 'red', marker='+')
ax.set_title('Average vs Strike Rate')
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')

plt.show()
```

```python
# two plots in single figure.

fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True)

ax[0].scatter(batsman['avg'], batsman['strike_rate'], color = 'red')
ax[1].scatter(batsman['avg'], batsman['runs'], color = 'lightgreen')

ax[0].set_title('Average vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Average vs Runs')
ax[1].set_ylabel('Strike Rate')

ax[1].set_xlabel('Average')
```

```python
# Four figures in a single figure. 

fig, ax = plt.subplots(nrows=2, ncols=2, figsize = (14, 8))

ax[0, 0].scatter(batsman['avg'], batsman['strike_rate'], color = 'red')
ax[0, 1].scatter(batsman['avg'], batsman['runs'], color = 'lightgreen')
ax[1, 0].hist(batsman['avg'])
ax[1, 1].hist(batsman['runs'])

ax[0, 0].set_title('Average vs Strike Rate')
ax[0, 1].set_title('Average vs Runs') 
ax[1, 0].set_title('Average') 
ax[1, 1].set_title('Runs')

ax[0, 0].set_xlabel('Average')
ax[0, 1].set_xlabel('Average') 
ax[1, 0].set_xlabel('Bins') 
ax[1, 1].set_xlabel('Bins')

ax[0, 0].set_ylabel('Strike Rate')
ax[0, 1].set_ylabel('Runs') 
ax[1, 0].set_ylabel('Average') 
ax[1, 1].set_ylabel('Runs')

plt.tight_layout()
```
