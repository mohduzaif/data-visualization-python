### Colored Scatterplots. 

```python
iris_df = sns.load_dataset('iris')
iris_df.sample(5)
```

```python
iris_df['species'] = iris_df['species'].replace({'setosa' : 0, 'versicolor' : 1, 'virginica' : 2})
```

```python
# cmap and alpha
plt.scatter(iris_df['sepal_length'], iris_df['petal_length'], c = iris_df['species'], cmap='jet', alpha=0.6)

plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()

plt.show()
```

### Subplots

```python
# how to make a subplots. 
fig, ax = plt.subplots(figsize = (10, 4))

ax.scatter(batsman['avg'], batsman['strike_rate'], color = 'red', marker='+')
ax.set_title('Average vs Strike Rate')
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')

plt.show()
```

```python
# two plots in single figure.

fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True)

ax[0].scatter(batsman['avg'], batsman['strike_rate'], color = 'red', marker = '+')
ax[1].scatter(batsman['avg'], batsman['runs'], color = 'lightgreen', marker = '*')

ax[0].set_title('Average vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Average vs Runs')
ax[1].set_ylabel('Strike Rate')

ax[1].set_xlabel('Average')
```

```python
# Four figures in a single figure. 

fig, ax = plt.subplots(nrows=2, ncols=2, figsize = (14, 8))

ax[0, 0].scatter(batsman['avg'], batsman['strike_rate'], color = 'red')
ax[0, 1].scatter(batsman['avg'], batsman['runs'], color = 'lightgreen')
ax[1, 0].hist(batsman['avg'])
ax[1, 1].hist(batsman['runs'])

ax[0, 0].set_title('Average vs Strike Rate')
ax[0, 1].set_title('Average vs Runs') 
ax[1, 0].set_title('Average') 
ax[1, 1].set_title('Runs')

ax[0, 0].set_xlabel('Average')
ax[0, 1].set_xlabel('Average') 
ax[1, 0].set_xlabel('Bins') 
ax[1, 1].set_xlabel('Bins')

ax[0, 0].set_ylabel('Strike Rate')
ax[0, 1].set_ylabel('Runs') 
ax[1, 0].set_ylabel('Average') 
ax[1, 1].set_ylabel('Runs')

plt.tight_layout()
```

### 3D Surface Plots. 

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
```

```python
xx, yy = np.meshgrid(x, y)
```

```python
# xx is repeation of x array in the rows to len(y) times. 
xx
```

```python
# yy is the repeation of columns to the len(x) times
yy
```

```python
z = xx**2 + yy**2
```

```python
# 3D plot_surface of fucntion x^2 + y^2

plt.figure()
ax = plt.subplot(projection = '3d')

plot_surface = ax.plot_surface(xx, yy, z, cmap = 'viridis')
ax.set_title('PLot of x^2 + y^2 function')

plt.colorbar(plot_surface)

plt.show()
```

```python
# one more function. 

z = np.sin(xx) + np.cos(yy)

plt.figure(figsize=(10, 6))
ax = plt.subplot(projection = '3d')

plot_surface = ax.plot_surface(xx, yy, z, cmap = 'jet')
plt.colorbar(plot_surface)

plt.show()
```

### Contour Plots or Contourf plot 
- contour plot shows the 3D plot into a 2d plot
- contourf shows plot with filled color.

```python
# 3D contour of fucntion x^2 + y^2

plt.figure()
ax = plt.subplot()

plot_surface = ax.contour(xx, yy, z, cmap = 'viridis')
ax.set_title('PLot of x^2 + y^2 function')

plt.colorbar(plot_surface)

plt.show()
```

```python
# 3D contour of fucntion x^2 + y^2
# blue color represent depth and yellow color represent peak.

plt.figure()
ax = plt.subplot()

plot_surface = ax.contourf(xx, yy, z, cmap = 'viridis')
ax.set_title('PLot of x^2 + y^2 function')

plt.colorbar(plot_surface)

plt.show()
```

```python
# contour plot. 
z = xx**2 + yy**2

plt.figure()
ax = plt.subplot()

plot_surface = ax.contour(xx, yy, z, cmap = 'viridis')
ax.set_title('PLot of x^2 + y^2 function')

plt.colorbar(plot_surface)

plt.show()
```

```python
# contourf plot.

z = xx**2 + yy**2

plt.figure()
ax = plt.subplot()

plot_surface = ax.contourf(xx, yy, z, cmap = 'viridis')
ax.set_title('PLot of x^2 + y^2 function')

plt.colorbar(plot_surface)

plt.show()
```

### Heatmap 
- used to compare two categorical columns.

```python
deliveries = pd.read_csv('../Datasets/IPL_Ball_by_Ball_2008_2022.csv')
deliveries.head()
```

```python
six_run_df = deliveries[deliveries['ballnumber'].isin([1, 2, 3, 4, 5, 6]) & (deliveries['batsman_run'] == 6)]
six_run_df.head()
```

```python
count_six_table = six_run_df.pivot_table(index='overs', columns='ballnumber', values='batsman_run', aggfunc='count')
count_six_table
```

```python
plt.figure(figsize = (20, 10))

plt.imshow(count_six_table)

plt.yticks(deliveries['overs'].unique(), list(np.arange(1, 21)))
plt.xticks(np.arange(0, 6), list(np.arange(1, 7)))

plt.colorbar()

plt.show()
```
