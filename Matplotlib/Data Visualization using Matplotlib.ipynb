## What is Matplotlib?

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is one of the most widely used Python packages for data visualization.

### Types of Data ?
- #### Numerical Data
        Data has numerical values like age, height, weight etc.
- #### Categorical Data
        Data that have some category like Gender, Branch, University etc.

`NOTE : ` 
- Analysis on single column is called Univariate Analysis.
- Analysis on Two column is called Bivariate Analysis.
- Analysis on more than two columns is called Multi-variate Analysis.

```python
# labels title. 
batsman = pd.read_csv('../Datasets/sharma-kohli.csv')

plt.plot(batsman['index'], batsman['RG Sharma'], color='red')
plt.plot(batsman['index'], batsman['V Kohli'], color='blue')

plt.title('Runs Comparision Rohit Sharma vs Virat Kohli')
plt.ylabel('Runs Scored')
plt.xlabel('Season')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6')

plt.title('Runs Comparision Rohit Sharma vs Virat Kohli')
plt.ylabel('Runs Scored')
plt.xlabel('Season')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2)

plt.title('Runs Comparision Rohit Sharma vs Virat Kohli')
plt.ylabel('Runs Scored')
plt.xlabel('Season')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3, marker='D', markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2, marker='o', markersize=10)

plt.title('Runs Comparision Rohit Sharma vs Virat Kohli')
plt.ylabel('Runs Scored')
plt.xlabel('Season')
```

```python
# legend -> location. 

plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3, marker='D', markersize=10, label = 'Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2, marker='o', markersize=10, label = 'Rohit')

plt.title('Runs Comparision Rohit Sharma vs Virat Kohli')
plt.ylabel('Runs Scored')
plt.xlabel('Season')

plt.legend(loc='upper left')
```

```python
# limiting axes  -----> To avoid the Outliers. 
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017,2019)
```

### Scatter Plots

image.png

- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation

```python
# Example 1
iq = [120, 100, 90, 95, 125, 115, 130, 140, 50, 60, 65]
marks = [90, 80, 70, 73, 95, 85, 99, 100, 50, 55, 63]

plt.scatter(iq, marks, color='red', marker='D')
```

```python
# Example 2 
x = np.linspace(-10, 10, 50)
y = 10 * x + 3 + np.random.randint(0, 200, 50)

plt.scatter(x, y, color='blue', marker='o')
```

```python
# plt.scatter on real pandas data. 
batsman = pd.read_csv('../Datasets/batter.csv')
batsman

batsman = batsman.head(50)
```

```python
plt.scatter(batsman['avg'], batsman['strike_rate'], color='red', marker='+')
plt.ylabel('Strike Rate')
plt.xlabel('Average')
plt.title('Analysis between Average and Strike Rate of Top 50 Players')

plt.grid()
plt.show()
```

```python
# size parameter.
tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size'] * 20)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Relation between Tip and Total Bill')
```

```python
# scatterplot using plt.plot
plt.plot(tips['total_bill'],tips['tip'],'o')
```

### `NOTE` 
--> plt.plot vs plt.scatter

- scatterplot using plt.plot is FASTER.
- scatterplot using plt.scatter is SLOWER.

### Bar chart

image.png

- Bivariate Analysis
- Numerical (Aggregate Function) vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink'] 

plt.bar(colors, children, color='green')
```

```python
# horizontal bar chart. 
plt.barh(colors, children)
```

```python
# bar chart using the real data. 
batters = pd.read_csv('../Datasets/batsman_season_record.csv')
batters
```

```python
plt.bar(batters['batsman'], batters['2015'])
plt.xlabel('Batsman')
plt.ylabel('Runs in 2015 Season')
```

```python
# Way of doing the task in matplotlib for multiple bar in single chart.
# Multiple Bar charts. 
plt.bar(np.arange(batters.shape[0]) - 0.2, batters['2015'], width=0.2, color='red')
plt.bar(np.arange(batters.shape[0]), batters['2016'], width=0.2, color='yellow') 
plt.bar(np.arange(batters.shape[0]) + 0.2 , batters['2017'], width=0.2, color='green')

# Replace the x-axis to other label name. 
plt.xticks(np.arange(batters.shape[0]), batters['batsman'])

plt.show()
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors, children)
```

```python
# Remove the above problem. 
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors, children, color='black')

plt.xticks(rotation = 'vertical')
plt.show()
```

```python
# stack bar chart. 
plt.bar(batters['batsman'], batters['2015'], label='2015')
plt.bar(batters['batsman'], batters['2016'], bottom=batters['2015'], label='2016')
plt.bar(batters['batsman'], batters['2017'], bottom=(batters['2015'] + batters['2016']), label='2017')


plt.legend()
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
marks = [23,45,100,20,49]
subjects = ['English','Science','Maths','SST','Hindi']
plt.pie(marks, labels=subjects)

plt.show()
```

```python
# real data. 
match_score = pd.read_csv('../Datasets/gayle-175.csv')
match_score
```

```python
plt.pie(match_score['batsman_runs'], labels=match_score['batsman'], autopct='%.2f%%', explode=[0.2,0,0,0,0,0.1])
plt.show()
```
