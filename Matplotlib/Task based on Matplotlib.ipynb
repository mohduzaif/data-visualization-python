### **`Problem-1:`** Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China. 

```python
polluted_cities = pd.read_csv('../Datasets/List of most-polluted cities by particulate matter concentration - List of most-polluted cities by particulate matter concentration.csv')
polluted_cities.head()
```

```python
temp_df = polluted_cities.query('Country == "Iran"')
iran_series = temp_df.groupby('Year')['PM2.5'].sum()
iran_series = iran_series.reset_index()
```

```python
china_series = polluted_cities.query('Country == "China"').groupby('Year')['PM2.5'].sum().reset_index()
china_series
```

```python
plt.plot(iran_series['Year'], iran_series['PM2.5'], label = 'Iran', linestyle = 'dotted', linewidth = 4)
plt.plot(china_series['Year'], china_series['PM2.5'], label = 'China', linestyle = 'dashed', linewidth = 4)

plt.xlabel('Year')
plt.ylabel('PM2.5')
plt.title('Year vs PM2.5')

plt.legend()
plt.grid(True)

plt.xticks(polluted_cities['Year'].unique())
plt.show()
```

### **`Problem-3:`** Draw a scatter plot where x-axis represents "PM2.5" and y-axis represents "PM10" for two countries Poland and Chile.

### `Problem-9` Show Quarter wise Sales data for all products as multi Bar chart.


```python
company_sales['date'] = pd.to_datetime(['2020-{}-01'.format(month) for month in company_sales['month_number']])
```

```python
# Drop 'date' column before grouping and summing
final_df = company_sales.drop(columns=['date']).groupby(company_sales['date'].dt.quarter).sum()
final_df
```

```python
i = -1
for col in final_df.columns[1:7]:
  plt.bar(final_df.index + i,final_df[col],width=0.15,label=col)
  i = i - 0.15

plt.xticks(final_df.index-1.4,final_df.index)
plt.xlabel('Product')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

### `Problem-10` Plot Stacked Bar chart quarter wise for each product.

```python
all_cols = []

for col in final_df.columns[1:7]:
  if len(all_cols) == 0:
    plt.bar(final_df.index,final_df[col],label=col)
  else:
    plt.bar(final_df.index,final_df[col],bottom=sum(all_cols),label=col)
  all_cols.append(final_df[col])

plt.xticks(final_df.index - 0.02, final_df.index)
plt.legend()
plt.show()
```
