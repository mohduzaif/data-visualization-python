### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# scatter plot -> axes level function
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', style='time', size='size')

plt.show()
```

```python
# relplot -> figure level -> square shape
# scatter using relplot -> size and hue
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind='scatter', hue='sex', style='time', size = 'size')
plt.show()
```

```python
# line plot. 
gapminder = px.data.gapminder()
gapminder.head()
```

```python
temp_df = gapminder[gapminder['country'] == 'India']
temp_df
```

```python
# lineplot -> axes level function.
sns.lineplot(data = temp_df, x = 'year', y = 'lifeExp')
```

```python
# relplot for lineplot -> this is also a figure level function. 
sns.relplot(data=temp_df, x = 'year', y = 'lifeExp', kind='line')
```

```python
temp_df = gapminder[gapminder['country'].isin(['India', 'Pakistan', 'China'])]
temp_df
```

```python
# using relplot which is a figure level function. 
sns.relplot(kind = 'line', data=temp_df, x = 'year', y = 'lifeExp', hue = 'country')
```

```python
sns.lineplot(data=temp_df, x = 'year', y = 'lifeExp', hue='country')
```

```python
temp_df = gapminder[gapminder['country'].isin(['India', 'Haiti', 'Zimbabwe'])]
temp_df
```

```python
sns.relplot(kind='line', data=temp_df, x = 'year', y = 'lifeExp', hue='country', style = 'continent', size='continent')
```

```python
# facet plot -> figure level function -> work with relplot

# it will not work with scatterplot and lineplot
```

```python
sns.relplot(kind='scatter', data=tips, x = 'total_bill', y = 'tip', col = 'sex', row='smoker')
```

```python
# col wrap
sns.relplot(data=gapminder, x='lifeExp', y='gdpPercap', kind='scatter', col='year')
```

```python
# col wrap -> manage by this parameter called col_wrap. 
sns.relplot(data=gapminder, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=3)
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot
```

```python
tips.head()
```

```python
sns.histplot(data = tips, x = 'total_bill')
plt.show()
```

```python
# bins. 
sns.histplot(data = tips, x = 'total_bill', bins=20)
plt.show()
```

```python
# element. 
sns.histplot(data = tips, x = 'total_bill', hue='sex', element='step', bins=20)
plt.show()
```

```python
# distplot. 
sns.displot(kind='hist', data=tips, x = 'total_bill', bins = 20, hue='sex', element = 'step')
```

```python
# Its also possible to visualize the distribution of a categorical variable using the logic of a histogram. 
# Discrete bins are automatically set for categorical variables

# countplot

sns.displot(kind='hist', data=tips, x = 'day')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x='age', kind='hist',element='step',hue='sex')
```

```python
# faceting using col and row -> not work on histplot function. 
sns.displot(kind='hist', data=tips, x = 'tip', col='sex', row='smoker')
```

```python
# kdeplot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# hue -> fill
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,height=10,aspect=2)
```

```python
# Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot 
# and then shows the count of observations within each rectangle with the fill color

# sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
# Heatmap

# Plot rectangular data as a color-encoded matrix. 
temp_df = gapminder.pivot(index='country', columns='year', values='lifeExp')
temp_df
```

```python

# axes level function
plt.figure(figsize=(15, 15))
sns.heatmap(data=temp_df)
```

```python
# annot. 
temp_df = gapminder[gapminder['continent'] == 'Europe'].pivot(index = 'country', columns = 'year', values = 'lifeExp')
temp_df
```

```python
# cmap, linewidth

plt.figure(figsize=(15,15))
sns.heatmap(data=temp_df, annot=True, linewidths=0.5, cmap='viridis')
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.

iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[ : , [0, 1, 2, 3]])
plt.show()
```
