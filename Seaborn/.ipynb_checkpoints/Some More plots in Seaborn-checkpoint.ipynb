```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import datsets. 
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

## Categorical Plots

### Categorical Scatter Plot

- Stripplot
- Swarmplot

### Categorical Distribution Plots

- Boxplot
- Violinplot

### Categorical Estimate Plot -> for central tendency

- Barplot
- Pointplot
- Countplot

### Figure level function -> `catplot`

### Categorical Scatter Plot.
- It is used to plot a scatter plot between categorical and numerical columns.
- It give me a bivariate analysis.


  **Stripplot** works well form medium and larger datasets and **swarmplot** is good for small datasets.

```python
# stripplot -> this is a axes level fuunction.
sns.stripplot(data=tips, x = 'day', y = 'total_bill', hue='sex')
```

```python
# catplot -> figure level function. 
sns.catplot(kind='strip', data=tips, x = 'day', y = 'total_bill', hue = 'sex')
```

```python
# swarm plot --> axes level function. 
sns.swarmplot(data=tips, x = 'day', y = 'total_bill', hue='smoker')
```

```python
sns.catplot(kind='swarm', data=tips, x = 'day', y = 'tip', s = 20, hue = 'smoker')
```

```python
# jitter parameter is also used for both the plots. 
sns.stripplot(data=tips, x = 'day', y = 'total_bill', hue='sex', jitter=0.2)
```

### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

```python
# boxplot. --> axes level function. 
sns.boxplot(data=tips, x = 'day', y = 'total_bill')
```

```python
# Using catplot
sns.catplot(data=tips,x='day',y='total_bill',kind='box')
```

```python
# hue Parameter. 
# Using catplot
sns.catplot(data=tips,x='day',y='total_bill',kind='box', hue='smoker')
```

```python
# single boxplot -> numerical column. 
sns.boxplot(data=tips, y = 'total_bill', hue='sex')
```

```python
# single boxplot -> numerical column. 
sns.boxplot(data=tips, y = 'tip', hue='sex')
```

### Violinplot = (Boxplot + KDEplot)

```python
sns.violinplot(data=tips, x = 'sex', y = 'tip', hue = 'smoker')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin', hue='sex')
```

```python
# split Parameter. 
sns.catplot(data=tips,x='day',y='total_bill',kind='violin', hue='sex', split = True)
```

```python
# barplot. 
sns.barplot(data=tips, x = 'sex', y = 'total_bill', hue = 'smoker')
```

```python
# errorbar parameter
sns.barplot(data=tips, x = 'sex', y = 'total_bill', hue = 'smoker', errorbar=None)
```

```python
# estimator parameter. 
sns.barplot(data=tips, x = 'sex', y = 'total_bill', hue = 'smoker', estimator='min')
```

```python
# point plot. --> show the difference of the error between two or more category given in that column.
sns.pointplot(data=tips, x = 'sex', y = 'total_bill', errorbar=None, hue='smoker')
plt.show()
```

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars

```python
# countplot  --> It is same as the histogram, this also used to calculate the count of the category of that given column. 
```

```python
sns.countplot(data=tips, x = 'day', hue='smoker')
plt.show()
```

A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable

```python
# faceting using catplot. 
sns.catplot(kind='box', data=tips, x = 'day', y = 'total_bill', col='sex', row='time', hue='smoker')
```

### Regression Plots

- regplot
- lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
# regplot ---> This is also an axes plot function.

# hue parameter is not avilable in regplot.
sns.regplot(data=tips, x = 'total_bill', y = 'tip')

plt.show()
```

```python
# limplot is a figure level function. 
sns.lmplot(data=tips, x = 'total_bill', y = 'tip', hue='sex')
```

```python
# residplot  --> It will show the error of prediction did by linear regression.
sns.residplot(data=tips,x='total_bill',y='tip')
plt.show()
```

### A second way to plot Facet plots -> FacetGrid

```python
# figure level -> relplot -> displot -> catplot -> lmplot.
```

```python
sns.catplot(data=tips, x='sex', y='total_bill', kind='box', col='day', row='time', hue='smoker')
plt.show()
```

```python
g = sns.FacetGrid(data=tips, col='day', row='time', hue='smoker')
# g.map 
g.map(sns.boxplot, 'sex', 'total_bill')
```

### Plotting Pairwise Relationship (PairGrid Vs Pairplot)

- **Pairplot** does not give me some much flexibilty it will give me these only plots that are shown below.
- **PairGrid** provide me a great flexibilty i can change the diagonal plots, upper-diagonal plots and lower-diagonal plots.

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species')
```

#### PairGrid function

```python
# sns.PairGrid()
g = sns.PairGrid(iris)
```

```python
# sns.PairGrid()
g = sns.PairGrid(iris, hue = 'species')

# g.map()
g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag

# sns.PairGrid. 
g = sns.PairGrid(iris, hue='species')

# map_diag
g.map_diag(sns.boxplot)

# map_offdiag 
g.map_offdiag(sns.kdeplot)
```

```python
# map_lower and map_upper and map_diag. 

g = sns.PairGrid(iris, hue='species')

g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
g.map_upper(sns.scatterplot)
```

```python
# vars
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

### JointGrid Vs Jointplot.

```python
# JointPlot.
```

```python
sns.jointplot(data=tips, x = 'total_bill', y = 'tip', hue='smoker')
```

```python
# JointGrid
```

```python
g = sns.JointGrid(data=tips, x = 'total_bill', y = 'tip')
g.plot(sns.scatterplot, sns.kdeplot)
```

```python
g = sns.JointGrid(data=iris, x = 'sepal_length', y = 'sepal_width', hue = 'species')
g.plot(sns.kdeplot, sns.histplot)

plt.show()
```
